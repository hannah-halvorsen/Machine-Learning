Do most stations naturally have enough open docks and bikes for people to pick up a bike or drop one off at their choosing? Or do stations struggle to meet demand for open docks and/or available bikes. Do some stations have more redistribution needs than others? How does the time of day, day of week, and month of year play into that? Is this a demand that is growing? 



### Conclusion
The initial EDA plots show there is a constant demand on the Blue Bike company to redistribute their bikes because the bikes are not used in a convenient pattern that would allow there to always be enough open docks and bikes at every station. We can see this by looking at the EDA plots that show there are some stations that need thousands of bikes per year to be restocked to them and there are other stations that have too many bikes being docked at them so bikes need to be taken away. The plots show that this demand to either refill or remove bikes changes vastly depending on the time of day, day of week, month, and year. And overall, that mainentanence demand is increasing each year. This project aims to predict how many bikes within a healthy confidence interval will need to be removed or replaced at a station over the year, month, day, and time of day. This will allow Blue Bikes to know how much staff they need to have at any given time and what stations they need to go to on any given morning or afternoon to drop off/pick up bikes. With successful forcasting, Bluebikes can get ahead of this demand and make sure customers show up to the Blue Bike stations and have a place to dock a bike they are done with or pick up a bike that they need to get where they are going. Additionally, if Blue bikes knows about how many bikes will need to be redistributed where, they could effectively set up routes for their drivers that pick up at one station that is going to be too full soon and drop those same bikes off at a station that is about to be empty.

## Create a categorical column to handle time of day to be used to analyze blue bike traffic patterns 
Rush hour times are based off of Boston Metropolitan Planning Organization https://www.ctps.org/subjects/traffic-volumes

## Description of bins
### morningrush_hour (6:00 am - 10:00 am): 
This category encompasses the early morning hours, specifically from 6:00 am to 10:00 am.

### midday (10:01 am - 3:00 pm): 
The midday category includes the time range from 10:01 am to 3:00 pm, covering the late morning to early afternoon hours.

### eveningrush_hour (3:01 pm - 7:00 pm): 
This category spans the late afternoon to early evening hours, from 3:01 pm to 7:00 pm.

### late_night and night (7:01 pm - 11:59 pm, and 12:00 am - 5:59 am): 
The late-night category combines the time ranges from 7:01 pm to 11:59 pm and 12:00 am to 5:59 am, covering the evening to early morning hours.

```python
# Create a new column 'time of day'
bb_all['time of day'] = pd.cut(
    bb_all['starttime'].dt.hour,
    bins=[0, 6, 10, 15, 19, 23],
    labels=['night', 'morningrush_hour', 'midday', 'eveningrush_hour', 'late_night'],
    include_lowest=True
)
```

### Find the top ten stations in the year 2021 with the largest difference between the number of occurrences of start station IDs and end station IDs
This shows relevance for the project because in 2021 there were almost ten stations with 1000 more bikes being taken away from a station then returned, which suggests bluebike employees were redistributing bikes to these stations so that they wouldn't be empty and unable to meet demand. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming bb2021 contains the dataset for the year 2021

# Group by 'start station id' and 'end station id' and count occurrences
station_counts_start = bb2021.groupby('start station id').size().reset_index(name='start_count')
station_counts_end = bb2021.groupby('end station id').size().reset_index(name='end_count')

# Merge counts for start and end stations
station_counts = station_counts_start.merge(station_counts_end, left_on='start station id', right_on='end station id', how='outer')

# Fill NaN values with 0
station_counts = station_counts.fillna(0)

# Calculate the difference between start station count and end station count
station_counts['difference'] = station_counts['start_count'] - station_counts['end_count']

# Sort by the difference in descending order
top_ten_stations = station_counts.sort_values(by='difference', ascending=False).head(10)

# Create a bar plot for the top ten stations
plt.figure(figsize=(12, 8))
sns.barplot(x='start station id', y='difference', data=top_ten_stations, palette='viridis')
plt.xlabel('Start Station ID')
plt.ylabel('Difference (Start Count - End Count)')
plt.title('Top Ten Stations with Largest Difference (Start Count - End Count) in 2021')

# Show the plot
plt.show()
```

## Next, I will test for stationarity of the residual series

```python
# Degine stationary test 

def test_stationarity(series, regression_model='c', rounding=3):
    """Function to test the stationarity of time series using:
    1. Augmented Dicky Fuller test using AIC 
    2. KPSS test for constant
    
    Arguments:  
        Series: the time series to be evaluated.   
        regression_model: model to be used:   
            'c' - model with a constant   
            'ct' - model wiht a constant and trend  
        rounding: the number of digits to display in printed output.   
    """
    if regression_model not in ['c','ct']:    
        print('Error!: argument regression model must be one of {\'c\',\'ct\'}')
        return
    with warnings.catch_warnings(): 
        warnings.filterwarnings("ignore")
        kpss_test = kpss(series, regression=regression_model)
        ADF_test = adfuller(series, regression=regression_model, autolag="AIC")
    return [round(ADF_test[0],rounding), round(ADF_test[4]['5%'],rounding), round(ADF_test[1],rounding),
                                       round(kpss_test[0],rounding), round(kpss_test[1],rounding)]
def stationary_tests(series, names, regression_model='c', rounding=3):
    test_columns = ['ADF_statistic','ADF_conf_int','ADF_p_value','KPSS_statistic','KPSS_p_value']
    out = pd.DataFrame(index=names, columns=test_columns)    
    for ser, name in zip(series,names):  
        out.loc[name,:] = test_stationarity(ser, regression_model=regression_model, rounding=rounding)
    return out
```

## Interpreting the stationary test
### The Adf statistic: The ADF stat is -8.2, which is significantly less than the critical values, indicating the null hypothesis of non-stationarity can be rejected.

### The ADF confidence interval: the ADF stat is less than these critical values, so I can reject the null hypothesis. 

### ADF p_value: The p-value is 0, providing evidence against the null hypothesis. 

### KPSS statistic: The kpss stat is 0.139. The p-value is 0.1, indicating that I should not reject the null hypothesis of stationarity.

## Conclusion: The ADF test provides strong evidence that the time series is stationary.  The KPSS also leans towards stationarity. Therefore, I conclude that the time series is likely stationary. 

```python
daily_net_bikes = daily_net_bikes.fillna(daily_net_bikes.mean())
daily_net_bikes_clean = daily_net_bikes[12:]
stationary_tests([daily_net_bikes_clean], ["Seasonal diff of log net bikes"])
```

```python
daily_net_bikes = daily_net_bikes.fillna()
best_model = pm.auto_arima(daily_net_bikes, 
                             max_p=3, max_q=3, 
                             max_P=3, max_Q=3, 
                             m=12, seasonal=True,
                             d=1, D=1, trace=True,
                             information_criterion = 'bic',
                             error_action='ignore',  # don't want to know if an order does not work
                             suppress_warnings=True,  # don't want convergence warnings
                             stepwise=True)  # set to stepwise
```

 # The optimal model is SARIMA (2,1,1)(3,1,1)12
 

```python
best_model.summary()
```

### Examination of the summary of the model 
- All of the model coefficients are significant and with relatively small standard errors. indicating the fit of the model is well-determined.
- The Ljung-Box test indicates that there is no serial correlation in the residuals. In other words, there is no predictive information not accounted for.
- The heteroskedasticity test suggests some level of heteroskedasticity, but the variance is not highly varying.
- The Jarque-Bera test indicates there is some significant skew or krutosis. This indicates strong evidence against the null hypothesis of normality in the residuals

# Forcasting with SARIMA 

### Next I will use the selected SARIMAX model to predict the time series values on the year 2022. I will make predictions using the model and see how it compares to the actual values 

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
SARIMAX_model = SARIMAX(daily_net_bikes, order=(2,1,1), seasonal_order=(3,1,1,12)).fit()
prediction = SARIMAX_model.get_forecast(365, dynamic=True).summary_frame()
prediction
```

```python
# Prepare actual data to be compared against the forcast data
# Combine all yearly dataframes into one list
bb_actual = [bb2015, bb2016, bb2017, bb2018, bb2019, bb2020, bb2021, bb2022]

# Concatenate yearly dataframes into one big dataframe
bb_actual = pd.concat(bb_actual, ignore_index=True)

# Convert 'start station id' and 'end station id' to numeric
# Concatenate yearly dataframes into one big dataframe
bb_actual['start station id'] = pd.to_numeric(bb_actual['start station id'], errors='coerce')
bb_actual['end station id'] = pd.to_numeric(bb_actual['end station id'], errors='coerce')

# Create a new column 'net_bikes' representing the surplus or need of bikes at each start station
bb_actual['net_bikes'] = abs(bb_actual['start station id'] - bb_actual['end station id'])

# Convert 'starttime' to datetime format
bb_actual['starttime'] = pd.to_datetime(bb_actual['starttime'])

# Set 'starttime' as the index for time series analysis
bb_actual.set_index('starttime', inplace=True)

# Resample the data by day and calculate the sum of net bikes for each day
daily_net_actual = bb_actual['net_bikes'].resample('D').sum()

# add a daily time series index and examine the head and tail of the data frame.
daily_net_actual.index = pd.date_range(start = '1-1-2015', end = '12-31-2022', freq = 'D')
print(daily_net_actual.head())
print(daily_net_actual.tail())

#convert to log
daily_net_actual = np.log(daily_net_actual)
daily_net_actual = daily_net_actual.diff(12)
daily_net_actual = daily_net_actual.dropna()
```
